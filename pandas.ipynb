## creating dataframe using list :

```python
import pandas as pd
import numpy as np
```

```python

data_list = [['Rahul', 29, 'Male'], ['Sahil', 24, 'Male'], ['Pratima', 24, 'Female']]
df_list = pd.DataFrame(data_list, columns=['Name', 'Age', 'Gender'])

display(df_list)
```

## creating dataframe using dictionary

```python
#list
names=['Rahul', 'Sahil', 'Pratima']
ages=[29, 24, 24]
genders=['Male', 'Male', 'Female']
#creating datframe
data_dict = {'Name':names, 'Age': ages, 'Gender':genders}
df_dict = pd.DataFrame(data_dict)

display(df_dict)
```

```python
data_numpy = np.array([['Rahul', 29, 'Male'], ['Sahil', 24, 'Male'], ['Pratima', 24, 'Female']])
df_numpy = pd.DataFrame(data_numpy, columns=['Name', 'Age', 'Gender'])

display(df_numpy)
```
